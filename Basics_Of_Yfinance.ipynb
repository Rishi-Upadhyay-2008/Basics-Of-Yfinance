```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

stocks =['JPM','GLD']

for stock in stocks :
  t = yf.download(stock, progress=False, auto_adjust=True)
  op = t['Open'].iloc[-5].item()
  cp = t['Close'].iloc[-5].item()
  if t.empty:
    print(stock,'no data')
    continue
  if cp>op :
      print(stock,'bull')
  else :
      print(stock,'bear')
```

```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import time

stocks = ['JPM','GLD','ORCL','^GSPC']
SL = []
FL = []


for stock in stocks :
  #print("Downloading:",stock)
  t = yf.download(stock, progress=False, auto_adjust=True)

  # Check if the DataFrame is empty before trying to access prices
  if t.empty:
    print(stock, 'no data available')
    continue

  # Access the last available open and close prices as scalar values
  InitalPrice = t['Close'].iloc[-20].item()
  FinalPrice = t['Close'].iloc[-1].item()

  if FinalPrice > InitalPrice :
    print(stock,'bullish',"Initial:",round(InitalPrice,3),'Final:',round(FinalPrice,3))
  else:
      print(stock,'bearish','Initial:',round(InitalPrice,3),'Final:',round(FinalPrice,3))
  SL.append(InitalPrice)
  FL.append(FinalPrice)

lable = [stocks[0],stocks[1],stocks[2],stocks[3]]
value = [SL[0],SL[1],SL[2],SL[3]]
value2 = [FL[0],FL[1],FL[2],FL[3]]
explode = (0.2,0.3,0.1,0)
plt.pie(value, labels = lable, autopct = '%1.1f%%', explode = explode)
plt.show()
```
